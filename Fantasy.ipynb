```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

matrix = df.as_matrix()

j = matrix[0]
s = matrix[1]
c = matrix[2]
l = matrix[3]

jc = np.cumsum(j[1:])
sc = np.cumsum(s[1:])
cc = np.cumsum(c[1:])
lc = np.cumsum(l[1:])

plt.figure(figsize=(15,15))
plt.plot(jc, linewidth=0.5, marker=".")
plt.plot(sc, linewidth=0.5, marker=".")
plt.plot(cc, linewidth=0.5, marker=".")
plt.plot(lc, linewidth=0.5, marker=".")
plt.ylabel('Points')
plt.show()
```
